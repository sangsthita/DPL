```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img,img_to_array
```

```python
#define the data augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=20,          #randomly rotate
    width_shift_range=0.1,      #Shift image width
    height_shift_range=0.1,     #shift image height
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)
```

```python
#load the original image
img_path=r"/content/download.jpeg"
img=load_img(img_path)
x=img_to_array(img)
```

```python
#reshape the image to a 4D tensor
img = np.expand_dims(x, axis=0)
```

```python
#Generate augmented images
augmented_images=[]
for batch in datagen.flow(img,batch_size=1):
    augmented_images.append(batch[0].astype(np.uint8))
    if len(augmented_images)>=5:
        break                 #stop the loop after generating 5 augmented images
```

```python
#Display the original and augmented images
plt.figure(figsize=(15,5))
for i,aug_img in enumerate(augmented_images,start=1):
      plt.subplot(1,5,i)
      plt.imshow(aug_img)
      plt.title(f'Augmented Image {i}')
      plt.axis('off')
plt.show()
```
